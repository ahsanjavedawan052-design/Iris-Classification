**Environment Setup & Data Loading**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

# Load dataset via URL
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(url, names=columns)

# Preview the data
df.head()
```

**Exploratory Data Analysis (EDA)**


```python
# Check distribution
print(df['species'].value_counts())

# Visualizing relationships
sns.pairplot(df, hue='species', markers=["o", "s", "D"])
plt.show()
```

**Data Preprocessing**

```python
# Split into Features and Target
X = df.drop('species', axis=1)
y = df['species']

# Split into Training (80%) and Testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

**Modeling (K-Nearest Neighbors)**

```python
# Initialize the model (using k=3 neighbors)
knn = KNeighborsClassifier(n_neighbors=3)

# Train the model
knn.fit(X_train, y_train)
```

**Evaluation & Prediction**

```python
# Make predictions
predictions = knn.predict(X_test)

# Evaluate performance
print(f"Accuracy: {accuracy_score(y_test, predictions) * 100}%")
print("\nClassification Report:\n", classification_report(y_test, predictions))

# Test with a new, unseen data point
new_flower = [[5.1, 3.5, 1.4, 0.2]] # Example morphological features
prediction = knn.predict(new_flower)
print(f"Predicted Species: {prediction[0]}")
```
